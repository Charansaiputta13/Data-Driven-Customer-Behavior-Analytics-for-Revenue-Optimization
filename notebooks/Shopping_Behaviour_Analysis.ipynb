```python
import types
import pandas as pd

def find_problem_cells(df):
    problems = []
    for col in df.columns:
        for i, val in df[col].items():
            # callable covers functions, bound methods, lambdas, etc.
            if callable(val):
                problems.append((col, i, "callable"))
                break
            # pandas objects inside a cell (DataFrame/Series) are also not serializable
            if isinstance(val, (pd.Series, pd.DataFrame)):
                problems.append((col, i, "pandas-obj"))
                break
            # weird types named 'method' sometimes appear; check type name too
            if type(val).__name__ == "method":
                problems.append((col, i, "method-type"))
                break
    return problems

problems = find_problem_cells(df)
print("Found problem columns (col, first_row_index, reason):", problems)
```

```python
import pandas as pd
import numpy as np

# mark any callables as missing
mask_callable = df['review_rating'].apply(lambda x: callable(x))
print("callable rows:", df.index[mask_callable].tolist())  # just to confirm

# Replace callables with NA
df.loc[mask_callable, 'review_rating'] = pd.NA

# Convert column to numeric (coerce invalid -> NaN)
df['review_rating'] = pd.to_numeric(df['review_rating'], errors='coerce')

print(df['review_rating'].dtype)
print("NaNs after coercion:", df['review_rating'].isna().sum())
```
